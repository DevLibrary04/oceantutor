```python
import pandas as pd
import json
from pathlib import Path
```

```python
# 디렉토리 순회
path = Path('C:/Users/temp/Downloads/해기사기출DB(2021-2023)')
for questionfolder in path.glob('*/*'):
    json_file_path = questionfolder / f"{questionfolder.name}.json"
    if not json_file_path.exists():
        print(f'{json_file_path} 존재하지 않음')
        continue
    with open(json_file_path, 'r', encoding='utf-8') as f:
        json_data = json.load(f)
        try:
            df = pd.json_normalize(
                json_data,
                record_path=['subject', 'type', 'questions'],
                meta=[
                    ['subject', 'name'],
                    ['subject', 'year'],
                    ['subject', 'inning'],
                    ['subject', 'type', 'string']
                ],
                errors='ignore' 
            )
            #do something with df
        except Exception as e:
            print(f"오류: {e}")
```
